## 2. Load Data <a id='2.loa'></a>

```python
def load_data_in_datasets(folderName, fileType):
    data_dir = './' + folderName
    data_dir = pathlib.Path(data_dir)
    image_count = len(list(data_dir.glob('*/*.' + fileType )))
    class_names = np.array([item.name for item in data_dir.glob('*')])
    
    #file_pattern = np.array([ folderName + '\\*\\' + item.name for item in data_dir.glob('*/*.' + fileType )])   
    file_pattern = tf.io.gfile.glob('./Data/*/*')
                            
    # note - https://github.com/tensorflow/tensorflow/issues/38879, abandoning the use of list_files due to non-deterministic random shuffle
    #list_ds = tf.data.Dataset.list_files(str(data_dir/'*/*'), shuffle=True)
    '''
    This is working as expected. repeat will re-execute the upstream computation for every repetition. 
    In the case shuffle=True different order of shuffling will be used (since the reshuffle_each_iteration argument of shuffle defaults to True).
    Since the reshuffle_each_iteration argument of shuffle is not exposed via list_files, you can either: 
    a) replace the use of list_files(pattern, shuffle=True) with from_tensor_slices(globbed_pattern).shuffle(num_files, reshuffle_each_iteration=False) 
    or b) add cache() after your take transformation.
    '''
    list_ds = tf.data.Dataset.from_tensor_slices(file_pattern).shuffle(SHUFFLE_BUFFER_SIZE, reshuffle_each_iteration=False, seed=13)    
    return list_ds, class_names, image_count


def get_label(file_path):
  # convert the path to a list of path components
  parts = tf.strings.split(file_path, os.path.sep)
  # The second to last is the class-directory
  return parts[-2] == class_names

def decode_img(img):
  # convert the compressed string to a 3D uint8 tensor
  img = tf.image.decode_png(img, channels=3)
  # Use `convert_image_dtype` to convert to floats in the [0,1] range.
  img = tf.image.convert_image_dtype(img, tf.float32)
  # resize the image to the desired size.
  return tf.image.resize(img, [imgHeight, imgWidth])

def process_path(file_path):
    label = get_label(file_path)
    # load the raw data from the file as a string
    img = tf.io.read_file(file_path)
    
    img = decode_img(img)
    return img, label

def process_path_with_filename(file_path):
    label = get_label(file_path)
    # load the raw data from the file as a string
    img = tf.io.read_file(file_path)
    
    img = decode_img(img)
    return img, label, file_path
```

### 5.21 Build and compile VGG16_MODEL <a id='5.21bui'></a>

```python
def VGG16_model():
    
    tf.keras.backend.clear_session()
    
    opt = tf.keras.optimizers.Adam()
    lossFunction = 'categorical_crossentropy'
    metrics = ['acc']

    VGG16_MODEL = tf.keras.applications.VGG16(input_shape=imgShape,
                                              include_top=False,
                                              weights='imagenet')
    VGG16_MODEL.trainable=False

    global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
    prediction_layer = tf.keras.layers.Dense(len(class_names), activation = predictionLayerActivation)

    model_VGG16 = tf.keras.Sequential([
        VGG16_MODEL,
        global_average_layer,
        prediction_layer])

    model_VGG16.compile(
        optimizer = opt,
        loss = lossFunction,
        metrics = metrics)
    
    return model_VGG16
```

### 5.26 Build and compile Custom Model <a id='5.26bui'></a>

[Go to 6.16 Fit and save Custom Model function](#6.16fit)<br>

```python
def custom_model(init_mode='he_uniform', activation='relu', dropout_rate=0.1, weight_constraint=0, neurons=32, learn_rate=0.001, momentum=0.9): #, optimizer='SGD'
    
    tf.keras.backend.clear_session()
    
    #opt = tf.keras.optimizers.SGD(lr=learn_rate, momentum=momentum)
    opt = tf.keras.optimizers.Adam(lr=learn_rate)
    
    lossFunction = 'categorical_crossentropy'
    metrics = ['acc']
      
    model = Sequential()
    model.add(Conv2D(neurons, (3, 3), activation=activation, kernel_initializer=init_mode, input_shape=(imgSize, imgSize, 3), data_format = 'channels_last'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Dropout(dropout_rate)) # added dropout to prevent overfitting
    model.add(Flatten())
    model.add(Dense(100, activation=activation, kernel_initializer=init_mode))
    model.add(Dense(len(class_names), activation=predictionLayerActivation))
    
    # compile model
    model.compile(
        optimizer = opt, 
        loss = lossFunction, 
        metrics = metrics)
    
    model.summary()
    return model
```

```python
def InceptionResNetV2_model():
    
    tf.keras.backend.clear_session()
    
    opt = tf.keras.optimizers.Adam()
    lossFunction = 'categorical_crossentropy'
    metrics = ['acc']

    InceptionResNetV2_MODEL = tf.keras.applications.InceptionResNetV2(input_shape=imgShape,
                                                          include_top=False,
                                                          weights='imagenet')

    InceptionResNetV2_MODEL.trainable=False

    global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
    prediction_layer = tf.keras.layers.Dense(len(class_names) , activation = predictionLayerActivation)

    model_InceptionResNetV2 = tf.keras.Sequential([
        InceptionResNetV2_MODEL,
        global_average_layer,
        prediction_layer
    ])

    model_InceptionResNetV2.compile(
        optimizer = opt,
        loss = lossFunction,
        metrics = metrics)
    
    return model_InceptionResNetV2
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalised confusion matrix")
    else:
        print('Confusion matrix, without normalisation')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
def ROC_curve(y_test, y_score):
    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    colors = cycle(['red', 'green', 'orange'])
    # Plot of a ROC curve for a specific class
    for i, color in zip(range(n_classes), colors):
        plt.figure(figsize=(10,6))
        plt.plot(fpr[i], tpr[i], color=color, label='ROC curve (area = %0.2f)' % roc_auc[i])
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(str(class_names[i]) + ' - Receiver operating characteristic')
        plt.legend(loc="lower right")
        plt.show()
```

```python
def multi_ROC_curve(y_test, y_score, modelName):
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    # Compute micro-average ROC curve and ROC area
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
    
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

    # Then interpolate all ROC curves at this points
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

    # Finally average it and compute AUC
    mean_tpr /= n_classes

    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

    # Plot all ROC curves
    lw = 2
    plt.figure(figsize=(16,10))
    plt.plot(fpr["micro"], tpr["micro"],
             label='micro-average ROC curve (area = {0:0.2f})'
             ''.format(roc_auc["micro"]),
             color='deeppink', linestyle=':', linewidth=4)

    plt.plot(fpr["macro"], tpr["macro"],
         label='macro-average ROC curve (area = {0:0.2f})'
             ''.format(roc_auc["macro"]),
             color='navy', linestyle=':', linewidth=4)

    colors = cycle(['red', 'green', 'orange'])
    for i, color in zip(range(n_classes), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=lw,
            label='ROC curve of {0} (area = {1:0.2f})' 
            ''.format(str(class_names[i]), roc_auc[i]))
            #''.format(i, roc_auc[i]))

    plt.plot([0, 1], [0, 1], 'k--', lw=lw)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(modelName + ' - Receiver operating characteristic')
    plt.legend(loc="lower right")
    plt.show()
```

```python
fig, ax = plt.subplots(len(models), 2, figsize = (20, 70))
fig.subplots_adjust(left = 0.02, right = 0.98, wspace = 0.5)
plt.rcParams.update({'font.size': 16})

for i, k in enumerate(m):
    # Plot training & validation accuracy values
    ax[i][0].plot(m[i]['acc'], label = 'Training')
    ax[i][0].plot(m[i]['val_acc'], label = 'Validation')
    ax[i][0].set_title(models[i] + ' Model Accuracy')
    ax[i][0].set_ylabel('Accuracy')
    ax[i][0].set_xlabel('Epoch')
    ax[i][0].legend()
    ax[i][0].grid(color='g', linestyle='-', linewidth=0.1)

    # Plot training & validation loss values
    ax[i][1].plot(m[i]['loss'], label = 'Training')
    ax[i][1].plot(m[i]['val_loss'], label = 'Validation')
    ax[i][1].set_title(models[i] + ' Model Loss')
    ax[i][1].set_ylabel('Loss')
    ax[i][1].set_xlabel('Epoch')
    ax[i][1].legend()
    ax[i][1].grid(color='g', linestyle='-', linewidth=0.1)
    

plt.show()
```
